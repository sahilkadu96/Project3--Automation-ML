```python
class IFNepitope_gamma_score:
    def __init__(self):
        self.base_url = 'https://webs.iiitd.edu.in/raghava/ifnepitope/predict.php'
        
    def get_data(self, source_filepath):
        driver.get(self.base_url)
        
        epitopes_list = []
        gamma_scores = []
        
        #open & read fasta file line by line
        with open(source_filepath, 'r') as f:
            ep = f.readlines()
        
        #we will provide input as batches 
        for i in range(0, len(ep), 20):
            ep10 = ep[i:i+20]
            ep10_str = ''.join(ep10)       #convert  list into string to provide in textbox
        
            text = driver.find_element_by_xpath('/html/body/div/div/div[3]/div/div/form/table/tbody/tr[3]/td/textarea')
            text.send_keys(ep10_str)
        
            submit = driver.find_element_by_xpath('/html/body/div/div/div[3]/div/div/form/input[2]')
            submit.click()
        
            WebDriverWait(driver, 180).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="example"]')))
        
            epit = driver.find_elements_by_xpath('//tbody/tr/td[3]/a')
            score = driver.find_elements_by_xpath('//tbody/tr/td[6]')
        
            for e, s in zip(epit, score):
                epitopes_list.append(e.text)
                gamma_scores.append(float(s.text))
                
            driver.get(self.base_url)
            sleep(3)
            
        driver.close()
            
        df = pd.DataFrame({"Epitopes":epitopes_list, 'IFN_gamma_score': gamma_scores})
        return df
    
    def save_results(self, df, dest_filepath):
        df.to_csv(dest_filepath, index = False)
```

```python
source_filepath = r'C:\Users\Sahil\Omdena\Development of Vaccines for JC Virus\mhc1_epitopes_fasta.txt'
dest_filepath = r'C:\Users\Sahil\Omdena\Development of Vaccines for JC Virus\IFN_gamma_score_results.csv'
```

```python
ifn = IFNepitope_gamma_score()
data = ifn.get_data(source_filepath)
data
```

```python
ifn.save_results(data, dest_filepath)
```
