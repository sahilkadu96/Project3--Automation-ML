```python
import pandas as pd
```

```python
df1 = pd.read_csv('epitopesII (MHC 1) - epitopesII.csv')
df1
```

```python
epitope_seq = df1['epitope sequence'].values
```

```python
epitope_seq
```

```python
def to_fasta(epitope_seq, file_path):
    seq_name = 'seq'
    with open(file_path, 'w') as f:
        count = 0
        for i in range(0, len(epitope_seq)):
            f.write('>' + seq_name + str(count) + '\n' + epitope_seq[i] + '\n')
            count += 1
            
    f.close()
    
    
```

```python
to_fasta(epitope_seq, 'all_epitopes_fasta.txt')
```

```python
mhc1_epit = df1[df1['length'] == 9]['epitope sequence'].values
```
